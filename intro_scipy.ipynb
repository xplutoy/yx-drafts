# python 科学计算、机器学习、画图等

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

## 数据集、make_circles

```python
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split

X, y = make_circles(100, factor=0.2, noise=0.1)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap="seismic")
plt.show()
```

# Kernel ridge regression

```python
from sklearn.kernel_ridge import KernelRidge
from sklearn.metrics.pairwise import polynomial_kernel
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y)

krr = KernelRidge(alpha=1.0, kernel='polynomial')
krr.fit(X_train,y_train)
```

```python
# 测试数据
_, axs = plt.subplots(1, 2)
axs[0].scatter(X_test[:, 0], X_test[:, 1], c=y_test)
axs[1].scatter(X_test[:, 0], X_test[:, 1], c=krr.predict(X_test))
```
